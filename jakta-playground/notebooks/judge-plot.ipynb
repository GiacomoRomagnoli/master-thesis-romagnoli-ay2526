```kotlin
%use kandy
%use dataframe
```

```kotlin
val df = DataFrame.readJson("data/judge_result.json")
```

```kotlin
df.results[0]
```

```kotlin
val eval = df.results[0]
    .remove { reason }
    .add("Model") {
        val provider = (it["eval_id"] as String).split("/").first { true }
        when(provider) {
            "openai" -> "GPT 4.1"
            "deepseek" -> "Deepseek Chat V3"
            "anthropic" -> "Claude Sonnet 4"
            "google" -> "Gemini 2.5 Flash"
            else -> provider
        }
    }
```

```kotlin
eval.groupBy { Model }
    .aggregate { mean("score") into "score" }
    .sortByDesc { Model and score }
    .plot {
        bars {
            x(Model) {
                axis.name = "Model"
            }
            y(score) {
                axis.name = "Score"
                axis.min = 0.0
            }
            fillColor("Model")
        }
    }
    .save("judge-eval.svg")
```
